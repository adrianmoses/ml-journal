## Using TFDS with Keras Models

Here's how to load in trainging and test data for feeding into keras models

Note: `as_supervised` ensures we get tuples of (X, y)

```python
(training_images, training_labels), (test_images, test_labels) = tfds.as_numpy(tfds.load('fashion_mnist',
  split=['train','test'],
  batch_size=-1,
  as_supervised=True))
```

Batch datasets with `.batch()` call

```python
data = tfds.load('horses_or_humans', split='train', as_supervised=True)

train_batches = data.shuffle(100).batch(10)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(300, 300, 3)),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(train_batches, epochs=1)
```

```python
import tensorflow_addons as tfa

def augment_images(image, label):
  image = tf.cast(image, tf.float32)
  image = (image/255)
  image = tf.image.random_flip_left_right(image)
  image = tfa.image.rotate(image, 40, interpolation='NEAREST') # rotate and fill with nearest neighboars
  return image, label
```

## Using Custom Splits

Apply custom splits with the `split=` keyword arg

```python
def load_cats_vs_dogs(split_string):
  return tfds.load('cats_vs_dogs', split=split_string, as_supervised=True)


train_data = load_cats_vs_dogs('train[:80%]')

validation_data = load_cats_vs_dogs('train[80%:90%]')

test_data = load_cats_vs_dogs('train[-10%:]')
```
